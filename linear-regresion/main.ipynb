```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

# Generate Data

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x = np.array([i*0.1 for i in range(jumlah_data)])

# print(y)
print(y)
```

### 1. Plot Data

```python
plt.scatter(x,y)
plt.title("Generate data")
plt.xlabel("x Label")
plt.ylabel("y Label")
plt.show()
```

### 2. fungsi linier

```python
def linear(x,gradien):
    y = gradien*x
    return y
```

### mengitung persamaan linier awal

```python
x_prediksi = np.array([0,10])
m_awal = 5
y_prediksi = linear(x_prediksi,m_awal)

print(y_prediksi)
```

### Menghitung Prediksi

```python
m_prediksi = m_awal
m_list_prediksi = []
y_list_prediksi = []
x_list_prediksi = []
learning_rate = 0.1

for i in range(1,jumlah_data):
    y_prediksi = linear(x[i],m_prediksi)
    y_actual = y[i]

    error = y_actual - y_prediksi
    delta_m = learning_rate*error/x[i]
    m_prediksi = m_prediksi + delta_m

    m_list_prediksi.append(m_prediksi)
    y_list_prediksi.append(linear(np.array([0,10]),m_prediksi))
    x_list_prediksi.append(np.array([0,10]))

```

### Visualisasi Prediksi

```python
fig = plt.figure(figsize=(4,4))
line, = plt.plot([],[],"Red")


def animate(frame_num):
    x = x_list_prediksi[frame_num]
    y = y_list_prediksi[frame_num]
    line.set_data((x,y))
    return line


plt.scatter(x,y)
plt.axis([0,10,0,10])
plt.xlabel('x')
plt.ylabel('y')
anime = FuncAnimation(fig,animate,frames=100,interval=100,repeat=False)
plt.show()
```
